```python
image = plt.imread("P3.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
lighten = gray_image.copy()
for i in range(lighten.shape[0]):
    for j in range(lighten.shape[1]):
        lighten[i,j] = max(0, min(255, 
                             np.sin(lighten[i,j]/255/2*np.pi)**2*255  
                        ))
```

```python
plt.hist(image.ravel(), 256, [0,256], color="black");
```

```python
image = plt.imread("P3.jpg")
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_image, cmap="gray")
```

```python
plt.hist(gray_image.ravel(), 256, [0,256], color="black");
```

```python
bw_image = (gray_image>120).astype(np.uint8)
```

```python
plt.imshow(bw_image, cmap="gray")
```

```python
kernel = np.ones((3,3))
plt.figure()
erode = cv2.erode(bw_image, kernel)
plt.imshow(erode, cmap="gray")

plt.figure()
plt.imshow(cv2.dilate(erode, kernel), cmap="gray")
```

```python
plt.imshow(
    cv2.medianBlur(bw_image, ksize=5),
    cmap="gray"
)
```

```python
plt.imshow(
    cv2.medianBlur(lenna, ksize=15),
)
```

```python
kernel = np.array([[-1,-1,-1],
                   [-1,9,-1],
                   [-1,-1,-1]])

plt.imshow(
    cv2.filter2D(lenna, -1, kernel)
)
```

```python
kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])

plt.imshow(
    cv2.filter2D(lenna, -1, kernel)
)
```

```python
kernel = np.array([[-1,-1,-1],
                   [0,0,0],
                   [1,1,1]])

plt.imshow(
    cv2.filter2D(lenna, -1, kernel)
)
```

```python
plt.imshow(
    (cv2.Sobel(lenna, -1, 0,1)+cv2.Sobel(lenna, -1,1,0))//2+lenna//2
)
```
